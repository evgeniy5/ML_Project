```python
def get_capacity_by_name(station):
    stations_in_one_entry = station[1].split(';')
    colors = station[0].split(';')
    for i in subways.keys():
        if colors != [''] and i not in colors:
            continue
        st = subways[i]
        for j in st.keys():
            for k in stations_in_one_entry:
                if j == k:
                    return st[j]            
```
