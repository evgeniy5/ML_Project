```python
import numpy as np
import pandas as pd
import pickle
import re
```

```python
streets = pd.read_table('streets')
housenumbers = pd.read_table('housenumbers')
build_years = pd.read_csv('moscow-buildings.csv')
```

```python
years = []

for i in range(len(housenumbers)):
    maybe1 = ''
    maybe2 = ''
    street = streets['street'][i]
    housenumber = re.findall(r'\d+', housenumbers['number'][i])
    for j in xrange(len(build_years)):
        street_ = str(build_years['street_name'][j])[:-1]
        flag = False
        if street.find(street_) != -1:
            if build_years['building'][j] == build_years['building'][j]: #nan != nan
                housenumber_ = re.findall(r'\d+', build_years['building'][j])
            else:
                housenumber_ = []
            maybe1 = build_years['house_year'][j]
            if housenumber_ and housenumber:
                if housenumber_[0] == housenumber[0]:
                    maybe2 = build_years['house_year'][j]
                    if len(housenumber) > 1 and len(housenumber_) > 1:
                        if housenumber_[0] == housenumber[0]:
                            flag = True
                    else:
                        flag = True
            if flag:
                years.append(build_years['house_year'][j])
                break
       
    if len(years) < i + 1:
        if maybe2:
            years.append(maybe2)
        elif maybe1:
            years.append(maybe1)
        else:
            years.append('not found')

    if i % 500 == 0:
        print 'process is %d/21000' % i
```

```python
for i in range(len(years)):
    if years[i] == '\xd0\xbd.\xd0\xb4.' or years[i] == 'not found':
        years[i] = '0'
```
